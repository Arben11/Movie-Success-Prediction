## Handling missing values

#### categorise numerical vs object values

- object columns: null values can be dropped (too much work to manually search and fill)
- numerical columns: fill na values with appropriate statistical operation

Handling Missing Values:

When addressing missing values, it's important to categorize columns as either numerical or object (categorical) values.

Object Columns (Categorical):
For object columns, which typically contain categorical information, a pragmatic approach is to drop rows with null values. Manually searching and filling these nulls can be labor-intensive and may not yield meaningful insights due to the categorical nature of the data.

Numerical Columns:
In contrast, for numerical columns, which involve quantitative data, it's more appropriate to perform data imputation to fill in the missing values. This involves using appropriate statistical operations like mean, median, or mode to replace missing values. The choice of imputation method depends on the distribution of the data and the potential impact on subsequent analyses.

For instance, in numerical columns like 'num_critic_for_reviews' or 'budget', we can calculate the mean or median of the available data and fill in the missing values with these computed values. This approach maintains the integrity of the data and avoids loss of valuable information.

```python
# Calculate IQR
Q1 = data["imdb_score"].quantile(0.25)
Q3 = data["imdb_score"].quantile(0.75)
IQR = Q3 - Q1

# Set a threshold for IQR (e.g., 1.5)
iqr_threshold = 1.5

# Find data points outside the IQR range
outliers_iqr = data[(data["imdb_score"] < Q1 - iqr_threshold * IQR) | (data["imdb_score"] > Q3 + iqr_threshold * IQR)]
print("Number of outliers for IMDb_score:", outliers_iqr.shape[0])
```

#### Hypothesis Testing

```python
popu = df['title_year'].value_counts().reset_index().sort_values(by='title_year', ascending=False)
temp = []
for i in range(len(popu)):
    temp.append(df.title_year[[popu.index[i]]].values//1)
movies = []
for i in range(len(popu)):
    movies.append(temp[i][0])
popu = pd.DataFrame(zip(movies, popu.title_year), columns=['Year', 'likes'])
plt.figure(figsize=(10, 7))
plt.title('Popularity of movies over the years')
sns.lineplot(data=popu, x='Year', y='likes')
plt.xticks(rotation=90)
plt.ylabel('Facebook Likes')
plt.xlabel('Year of Release', fontsize=12)
plt.show
```

- From the plot we can infer that though the numebr of movies increased drastically in the late 2000s, the popularity of Movies peaked during the mid 2000s, while gradually decreasing after that on social Media platforms like facebook.

```python
from scipy import stats

pop_mean = popu.likes.mean()
t_value, p_value = stats.ttest_1samp(popu.likes, pop_mean)
one_tailed_p_value=float("{:.6f}".format(p_value/2)) # Since alternative hypothesis is one tailed, We need to divide the p value by 2.
print('Test statistic is %f'%float("{:.6f}".format(t_value)))
print('p-value for one tailed test is %f'%one_tailed_p_value)
alpha = 0.05
if one_tailed_p_value<=alpha:
    print('Conclusion','n','Since p-value(=%f)'%p_value,'<','alpha(=%.2f)'%alpha,'''We reject the null hypothesis H0. So we conclude that there is no significant mean difference in systolic blood pressure. i.e., μ = 165 at %.2f level of significance'''%alpha)
else:
    print('Conclusion','n','Since p-value(=%f)'%one_tailed_p_value,'>','alpha(=%.2f)'%alpha,'''\nWe do not reject the null hypothesis H0''')
```

- We Accept the NULL Hypothesis.
- It implies that we do not have any evidence that the popularity of movies increased with time.

#### 4.3.1 Checking for Outlieres

There are several ways to check for outliers in a dataset:

- Summary statistics: Another way to identify outliers is to calculate summary statistics such as the mean, median, standard deviation, and range of the data. Outliers can be identified as data points that are several standard deviations away from the mean or median, or as data points that fall outside the minimum and maximum range.

- Visualization: One of the most effective ways to identify outliers is to visualize the data using plots such as boxplots, histograms, or scatterplots. Outliers can be identified as points that are far away from the main cluster of data points.

- Quantile-based methods: The interquartile range (IQR) is a commonly used method for identifying outliers. It is defined as the difference between the third quartile (Q3) and the first quartile (Q1) of the data. Outliers can be identified as data points that fall outside the range Q1 - 1.5IQR to Q3 + 1.5IQR.

- Once outliers have been identified, the appropriate course of action depends on the nature of the data and the goals of the analysis. In some cases, outliers may need to be removed or corrected, while in other cases they may be genuine observations that need to be retained.

### Z-Score

- Another method of checking for outliers is to calculate the z-score for each observation. The z-score measures the number of standard deviations a data point is from the mean of the dataset. Observations with a z-score greater than 3 or less than -3 are often considered outliers.

- It is important to note that not all outliers are bad data points, and they may represent real observations that are important to understanding the phenomenon being studied. Therefore, it is important to carefully evaluate and understand the nature of any outliers before deciding how to handle them in statistical analyses.

```python
import numpy as np
import pandas as pd

def remove_outliers(df, cols):
    for col in cols:
        z_scores = np.abs((df[col] - df[col].mean()) / df[col].std())
        df = df[z_scores < 3]
    return df
```

Recursive Feature Elimination with Cross Validation

Recursive — involving doing or saying the same thing several times in order to produce a particular result or effect

Feature — individual measurable property or characteristic of a phenomenon being observed — an attribute in your dataset

Cross-Validation — a technique for evaluating ML models by training several ML models on subsets of the available input data and evaluating them on the complementary subset of the data. Use cross-validation to detect overfitting, ie, failing to generalize a pattern.

You will need to declare two variables — X and y where first represents all the features, and the second represents the target variable. Then you’ll make an instance of the Machine learning algorithm (In this case RandomForests). In it, you can optionally pass a random state seed for reproducibility. Now you can create an instance of RFECV.

```python
#Performing Recursive Feauture Elimation with Cross Validation
#Using Random forest for RFE-CV and logloss as scoring
from sklearn.feature_selection import RFECV
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import log_loss
clf_rf=RandomForestClassifier(random_state=0)
rfecv=RFECV(estimator=clf_rf, step=1,cv=5,scoring='neg_log_loss')
rfecv=rfecv.fit(X_train,y_train)
```
